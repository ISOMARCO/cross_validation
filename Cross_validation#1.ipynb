```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df_columns = df.columns
```

```python
for i in df_columns:
    if(df.dtypes[i] == 'object'):
        print(i)
```

```python
y = df['Bankrupt?']
```

```python
x = df.drop('Bankrupt?', axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
#SVC
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
```

```python
help(BaggingClassifier)
```

```python
model = SVC()
bagging = BaggingClassifier(base_estimator = model, random_state = 42)
```

```python
bagging.fit(x_train, y_train)
```

```python
pred = bagging.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
#Boosting
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
help(GradientBoostingClassifier)
```

```python
boosting = GradientBoostingClassifier(learning_rate = 0.01, random_state = 42) #decision tree
```

```python
boosting.fit(x_train, y_train)
```

```python
pred_boosting = boosting.predict(x_test)
```

```python
accuracy_score(y_test, pred_boosting)
```
